```python
import cv2
import numpy as np
import tensorflow as tf
```

```python
# Load your trained model
model = tf.keras.models.load_model("saved.keras")  # Update this with your model's path
```

```python
# Initialize video capture
video_path = "lemon.mp4"  # Update this with your video file path
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error opening video file")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame as needed (resize, normalization, etc.)
    processed_frame = cv2.resize(frame, (256, 256))  # Placeholder for preprocessing

    # Perform prediction
    input_data = np.expand_dims(processed_frame, axis=0)
    predictions = model.predict(input_data)
    predicted_class = np.argmax(predictions[0])
    confidence = np.max(predictions[0]) * 100

    # Get bounding box coordinates based on the prediction
    # These are example values; replace them with your logic based on the prediction
    x, y, w, h = 350, 200, 100, 100  # Example bounding box coordinates
    
    # Draw bounding box and label on frame
    class_names = ["diseased", "healthy"]  # Replace with your class names
    color = (0, 255, 0)  # Green color for the bounding box
    cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)  # Draw bounding box
    text = f"Prediction: {class_names[predicted_class]}, Confidence: {confidence:.2f}%"
    cv2.putText(frame, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Display the frame with prediction and bounding box
    cv2.imshow('Video Detection', frame)

    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
